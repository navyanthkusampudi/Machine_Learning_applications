#### 4. Changes in electronicConfiguration	
Extracting number of electrons from clectronic configuration. 
For example: '[Rn] 5f14 6d10 7s2 7p2' ==> {'s':2,'p':2,'d':10,'f':14}

```python
orbitals = list('spdf')
orbitals
```

```python
test_string =  ' [Rn] 5f14 6d10 7s2 7p2 '
```

```python
test_s1 = re.sub('\[.*]','',test_string) # ex: remove [RN]
test_s1
```

```python
# find the group according to the requirement 
group1 = '([0-9])'   # ex: 5
group2 = '([spdf])'  # ex: f
group3 = '([0-9]*)'  # ex: 14

searched_pairs = re.findall(group1+group2+group3, test_s1)
```

```python
searched_pairs
```

#### function to extract the dictionary of spdf electrons from a string

```python
def extract_spdf_elec(s):
    
    s1 = re.sub('\[.*]','',s) 
    group1 = '([0-9])'   # ex: 5
    group2 = '([spdf])'  # ex: f
    group3 = '([0-9]*)'  # ex: 14
    match_pairs = re.findall(group1+group2+group3, s1)
    
    return {b: int(c) for a,b,c in match_pairs}
```

```python
ex = extract_spdf_elec(test_string)
ex
```

```python
print(ex.get('s'))
print(ex.get('p'))
print(ex.get('d'))
print(ex.get('f'))
print(ex.get('l',0))
```

#### apply the function to all data points in dictionary

```python
df['electronicConfiguration_dict'] = df['electronicConfiguration'].map(lambda s: extract_spdf_elec(s))
```

```python
df[['electronicConfiguration', 'electronicConfiguration_dict']]
```

## Fitting the data to linear curve

```python
from sklearn.linear_model import LinearRegression
```
