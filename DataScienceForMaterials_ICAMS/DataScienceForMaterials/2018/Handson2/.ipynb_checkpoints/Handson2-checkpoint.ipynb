In order to remove them, let just drop out all data that outside of inter-quartile range. We will use built-in to dataframe function .quantile at two values 0.25 and 0.75 which corresponds to Q1 and Q3. 25%-quantile, for example, means such a value that 25% of data is below it

```python
quantile_df = df.quantile(q=[0.25,0.75]).T
quantile_df.columns=["Q1","Q3"]
quantile_df
```

calculate inter quantile range

```python
quantile_df["IQR"] = quantile_df["Q3"]-quantile_df["Q1"]
```

```python
quantile_df["Lower_bound"] = quantile_df["Q1"]-1.5*quantile_df["IQR"]
quantile_df["Upper_bound"] = quantile_df["Q3"]+1.5*quantile_df["IQR"]
```

```python
quantile_df
```

now we will iterate over the columns of original dataframe and set those values that are outside the [Q1-1.5IQR,Q3+1.5IQR] range to Not-A-Number (NaN)

```python
for col in df.columns:
    is_below_lower_bound_mask = df[col]<quantile_df.loc[col,"Lower_bound"]
    is_above_upper_bound_mask = df[col]>quantile_df.loc[col,"Upper_bound"]
    df.loc[ is_below_lower_bound_mask | is_above_upper_bound_mask, col]=np.NaN
```

A few comments on this complex expression:

Dataframe has a methods to access to its content by general index (which could be anything) with .loc and by integer index, like in arrays, with .iloc.

`quantile_df.loc[col,"Lower_bound"]`

means the value of `quantile_df` for index `col` at column "Lower_bound". It will be a single number.

Then we compare this single number to every element in column `df[col]`. It will give an array of booleans (True or False) - results of comparison, which indicates if the given value is below our lower bound. The same thing for upper bound. We stored this arrays in `is_below_lower_bound_mask` and `is_above_upper_bound_mask`.

Then we combined both of them with OR operator (|) (below OR above bounds), and used it as a filter for our original data frame, specify the current column `col` and assign to these elements NaN value:

`df.loc[ is_below_lower_bound_mask | is_above_upper_bound_mask, col] = np.NaN`

Finally, we drop out all rows with at least one NaN value.

```python
df.dropna(inplace=True)
```

We will write a helper function, that will take an sklearn regressor, and score it

```python
def my_scorer(regressor):
    """
    Score a given regressor on both train and test datasets,
    calculate metrics (RMSE, MAE, R2)
    and plot true-vs-predicted values.
    
    Parameters:
    -----------
    regressor - a sklearn-compatible regressor that has .predict method
    
    """
    pred_train = regressor.predict(X_train)
    pred_test = regressor.predict(X_test)
    print("Train data:")
    print("RMSE = ",np.sqrt(mean_squared_error(y_train,pred_train)))
    print("MAE = ",mean_absolute_error(y_train,pred_train))
    print("R2 = ",r2_score(y_train,pred_train))
    
    print("Test data:")
    print("RMSE = ",np.sqrt(mean_squared_error(y_test,pred_test)))
    print("MAE = ",mean_absolute_error(y_test,pred_test))
    print("R2 = ",r2_score(y_test,pred_test))
    
    plt.scatter(y_train,pred_train,label="train",color="blue", alpha=0.1)
    plt.scatter(y_test,pred_test,label="test",color="green", alpha=0.1)
    plt.plot([30,80],[30,80])
    plt.show()
```

A text inside triple quotes at the beginning of procedure body is a docstring. It will be shown if you press Shift+Tab or run a command name with question mark

```python
my_scorer?
```

```python
def train_test_r2_metrics(regressor):
    """
    Score a given regressor on both train and test datasets and
    calculate R2 metrics.
    

    Parameters:
    -----------
      regressor - a sklearn-compatible regressor that has .predict method
      
    Returns:
    --------
        r2_train, r2_test
    """
    pred_train = regressor.predict(X_train)
    pred_test = regressor.predict(X_test)    
    r2_train=r2_score(y_train,pred_train)
    r2_test = r2_score(y_test,pred_test)    
    return r2_train,r2_test
```

## Feed forward feature selection in linear models

Idea of feed-forward selection is to start from empty set of data-columns (Al, Ti, T, ...) and at each step to add the column that gives the best improvment of regression metric ($R^2$ for example) for linear regression.
Procedure stops when no further improvment is possible or all columns are included.

The Python code for this method was taken from:

https://planspace.org/20150423-forward_selection_with_statsmodels/

```python
import statsmodels.formula.api as smf
```

```python
def forward_selected(data, response):
    """Linear model designed by forward selection.

    Parameters:
    -----------
    data : pandas DataFrame with all possible predictors and response

    response: string, name of response column in data

    Returns:
    --------
    model: an "optimal" fitted statsmodels linear model
           with an intercept
           selected by forward selection
           evaluated by adjusted R-squared
    """
    remaining = set(data.columns)
    remaining.remove(response)
    selected = []
    current_score, best_new_score = 0.0, 0.0
    while remaining and current_score == best_new_score:
        scores_with_candidates = []
        for candidate in remaining:
            formula = "{} ~ {} + 1".format(response,
                                           ' + '.join(selected + [candidate]))
            score = smf.ols(formula, data).fit().rsquared_adj
            scores_with_candidates.append((score, candidate))
        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates.pop()
        if current_score < best_new_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            print("Added: ",best_candidate, " score: ",best_new_score)
            current_score = best_new_score
    formula = "{} ~ {} + 1".format(response,
                                   ' + '.join(selected))
    model = smf.ols(formula, data).fit()
    return model
```

The similar approach to feed-forward selection method is in a backward selection, with the only difference thatone starts from the full set of columns and eliminated them one-by-one:

```python
def backward_selected(data, response):
    """Linear model designed by backward selection.

    Parameters:
    -----------
    data : pandas DataFrame with all possible predictors and response

    response: string, name of response column in data

    Returns:
    --------
    model: an "optimal" fitted statsmodels linear model
           with an intercept
           selected by backward selection
           evaluated by adjusted R-squared
    """
    remaining = set(data.columns)
    remaining.remove(response)
    selected = list(remaining) #all features are selected
    current_score, best_new_score = 0.0, 0.0
    while remaining and current_score == best_new_score:
        scores_without_candidates = []
        for candidate in selected:
            current_without_candidate = set(selected)
            current_without_candidate.remove(candidate)
            formula = "{} ~ {} + 1".format(response,
                                           ' + '.join(current_without_candidate))
            score = smf.ols(formula, data).fit().rsquared_adj
            scores_without_candidates.append((score, candidate))
        scores_without_candidates.sort()
        best_new_score, best_candidate = scores_without_candidates.pop()
        if current_score < best_new_score:
            selected.remove(best_candidate)
            print("Removed: ",best_candidate, " score: ",best_new_score)
            current_score = best_new_score
    formula = "{} ~ {} + 1".format(response,
                                   ' + '.join(selected))
    model = smf.ols(formula, data).fit()
    return model
```

```python
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
from sklearn.externals.six import StringIO  

def visualize_tree(dtree):
    dot_data = StringIO()
    export_graphviz(dtree,out_file=dot_data,  
                    filled=True, rounded=True,
                    special_characters=True)
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
    return Image(graph.create_png())
```

Congratulations, you have zero train error and R2=1. But test error is very high, because decision tree have learnt train data "by heart" and thus, loose its generalization ability.
This is the major problem of ML - **overfitting**

Now, the depth of the tree is 

```python
tree_regressor.tree_.max_depth
```

In principle, such a tree could learn up to 2$^{29}$ data points by heart.

```python
2**29
```

In order to deal with overfitting, one has use **regularization** methods. They are depends on the ML algorithm. For the decision tree you can limit its max_depth, or consider another parameters from documentation.

```python
%%time

max_depth_list=[]
r2_train_lists=[]
r2_test_list=[]

for max_depth in range(1,30,1):
    print("Training with max_depth = ",max_depth)
    tree_regressor = DecisionTreeRegressor(max_depth=max_depth)
    tree_regressor.fit(X_train,y_train)
    r2_train,r2_test = train_test_r2_metrics(tree_regressor)
    r2_train_lists.append(r2_train)
    r2_test_list.append(r2_test)
    max_depth_list.append(max_depth)
```

```python
plt.plot(max_depth_list,r2_train_lists,label="train")
plt.plot(max_depth_list,r2_test_list,label="test")
plt.xlabel("max_depth")
plt.ylabel("R2")
plt.legend()
```

From this plot one can see, that best R2 value on test data is acheived at max_depth=11, whereas train score is monotonically improved. Max_depth is a **hyperparameter** of decision tree algorithm, because it determines the overall behaviour of fitting procedure.

Search of the optimal hyperparameters values is called **hyperoptimization** (https://en.wikipedia.org/wiki/Hyperparameter_optimization)

```python
tree_regressor = DecisionTreeRegressor(max_depth=11)
tree_regressor.fit(X_train,y_train)
my_scorer(tree_regressor)
```

## Ensemble methods: random forest

Ensemble methods combine the prediciton of **weak learner** (i.e. decision tree with small max_depth) 

```python
from sklearn.ensemble import RandomForestRegressor
```

A random forest is a meta estimator that fits a number of decision tree classifiers on various sub-samples (if bootstrapping is used) of the dataset and use averaging to improve the predictive accuracy and control over-fitting.  (http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html)

Already combining only *two* decision trees with max_depth=11 test score is slightly higher (0.7384 vs. 0.7378). Do not forget to set random_state to some number, to have a reproducible results.

```python
random_forest = RandomForestRegressor(n_estimators=2,max_depth=11,bootstrap=False, random_state=42,)
random_forest.fit(X_train, y_train)
my_scorer(random_forest)
```

by allowing bootstraping (https://en.wikipedia.org/wiki/Bootstrapping_(statistics)) we increase metric even more (0.7814 vs. 0.7384). This is the main idea of **ensemble learning - composition of predictions from independent models**

```python
random_forest = RandomForestRegressor(n_estimators=2,max_depth=11,bootstrap=True,random_state=42,)
random_forest.fit(X_train, y_train)
my_scorer(random_forest)
```

```python
random_forest = RandomForestRegressor(n_estimators=50,random_state=42, max_depth=11, n_jobs=-1,verbose=1)
random_forest.fit(X_train,y_train)
my_scorer(random_forest)
```

Idea is to stack different layers to build a NN.

Add a dense layers (a layer in which all neurons are connected to all neurons from previous layer) with different amount of neurons and different activation functions (https://en.wikipedia.org/wiki/Activation_function):
 - non-linear(tanh) at first hidden layer with 200 neurons

```python
nn = Sequential()
```

```python
nn.add(Dense(200,input_shape=(X_train.shape[1],), activation="tanh"))
```

```python
nn.add(Dense(100,activation="tanh"))
```

```python
nn.add(Dense(1,activation="linear"))
```

Now, we "compile" the model and set up a loss function (mean squared error) and optimizer to minimize the loss (Adam optimizer, https://keras.io/optimizers/#adam)

```python
nn.compile(optimizer="adam",loss="mean_squared_error")
```

see the summary of our model architecture

```python
nn.summary()
```
