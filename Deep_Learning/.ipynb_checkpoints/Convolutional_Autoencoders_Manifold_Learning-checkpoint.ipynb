The image is in 784 dimension vector space. This space is very high dimensional and most dimensionality reduction techniques try to exploit the assumption that not all of the dimension are needed to distinguish btw the digits.

**Where does this intuition comes from?**

To see this lets us generate a point uniformly at random in the unit 784 dimensional hypercube and see what it looks like. We pick random points in the hypercube simply because we normalized the pixel intesities.


```python
#generating a random 28 by 28 image:
rand_img = np.random.randint(0,255,(28,28))
rand_img = rand_img/255.0

show(rand_img)
```

This image doesnt make anysense to us. We can try to sample at random many times but unless we get extremely lucky all we will get is static and nothing resembling actual digit. This is good emperical evidence that the meanigful images are clustered in smaller dimensional subsets in the original 784 dimensional pixel space. This is called the manifol hypothesis. And the promise is that if we better understand the structure of the manifold we will have an easier time building machine learning.

Before we try to see how to figure out the manifold structure let us take a close look at the out space. For example what happens if we start at the point of adigit and start travellig in random direction ? will we get anymeaningful images?

```python
rand_direction = np.random.rand(28,28)
```

#### Moving in a random direction away from digit in the 784 dimesional images space:

```python
for i in range(16):
    plt.subplot(4,4,i+1)
    show(img+(i/4)*rand_direction)
    plt.xticks([])
    plt.yticks([])
```

We can see that as we move away from the digit, the images we encounter become less and less distinguishable. At first we cn still see the shape of of three but before we know it we are back to static land.

Perhaps a good analogy here is that of a solar system: The surface of our planets are th manifolds we are interested in, one for each digit. Now we say you are on the surface of earth which is 2-D manifold and you start moving in a random direction (asume gravity doesnt exist and we can go through solid objects). If you dont understand the structure of earth you will quickly find yourself  in space or inside the earth. But if you instead move within the local earth coordinates you will stay on the surface and get to see all the cool stuff.

There are however some differences: First of all we are much in higher dimensional space and we are not sure how many dimensions we need to capture the structure f digit subspaces. Secondly these subspaces could be really crazy looking (Think of a doughnut entangled in a wierd way). You could in fact get from one manifold to another without going into static space at all.

#### Learning  the Manifold with an auto encoder

Ok so how do we figure out what dimensions in the image space are important ? One option would be to hand engineer features - for example the mean of all pixels is probably a good feature to have. Other worthwile features would be the slant and vertical or horizontal symmetry. 

But we want to do machine learnig not hand craft features because we are lazy and machines tend to better capture important features in messy datasets. There are many ways to reduce the dimensionalty, here we will use Autoencoders. 

An Autoencoder is a feed-forward neural network who tries to learn a lower dimensional representation of our data. It does that by decreasing the dimension of the original image.

Since the auto encoder is forced to reconstruct the images from a smaller representation it disregards any variation that it doesnt find useful. 

'The important principle is that the autoencoder can afford to represent only the varialtions that are needed to reconstruct training examples. If the data generating sitribution concentrates near low-mdimensional manifold, this yields representations that implicitily capture a local coordinate systems for this manifold: only the variations tangent to the manifold around x need to correspond to changes in h=f(x)'

### Building Convolutional Auto encoder

```python
input_img = Input(shape=(28,28,1))
encoded = Convolution2D(32,(3,3),activation='relu')(input_img) #26
encoded = MaxPooling2D((2,2),padding='same')(encoded)  # 13
encoded = Convolution2D(32,(3,3),activation='relu')(encoded) # 11
encoded = MaxPooling2D((2,2),padding='same')(encoded) # 6
encoded = Convolution2D(32,(3,3),activation='relu')(encoded) # 3
encoded = MaxPooling2D((2,2),padding='same')(encoded) # 
encoded = Flatten()(encoded)
encoded = Dense(64,activation='relu')(encoded)
encoded = Dense(2)(encoded) # 2 here

decoded = Dense(64,activation='relu')(encoded)
decoded = Dense(128,activation='relu')(decoded)
decoded = Reshape((2,2,32))(decoded) # 2
decoded = UpSampling2D((2,2))(decoded) #4
decoded = Convolution2D(32,(3,3),padding='same',activation='relu')(decoded)
decoded = UpSampling2D((2,2))(decoded) #8
decoded = Convolution2D(32,(3,3),padding='same',activation='relu')(decoded)#8
decoded = UpSampling2D((2,2))(decoded) #16
decoded = Convolution2D(32,(3,3),activation='relu')(decoded)#14
decoded = UpSampling2D((2,2))(decoded) #28
decoded = Convolution2D(1,(3,3),activation='sigmoid',padding='same')(decoded)

autoencoder = Model(input=input_img, output=decoded)
```

```python
autoencoder.summary()
```

```python
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(X_train.reshape(X_train.shape[0],28,28,1),
                X_train.reshape(X_train.shape[0],28,28,1), 
                batch_size=128,nb_epoch=10, verbose=1)
```

```python
image= autoencoder.predict(X_train[200].reshape(1,28,28,1))
```

```python
# how well does the autoencoder decode ?
plt.subplot(2,2,1)
show(X_train[200])
plt.subplot(2,2,2)
image= autoencoder.predict(X_train[200].reshape(28,28,1))
show(image.reshape(28,28))

plt.subplot(2,2,3)
show(X_train[300])
plt.subplot(2,2,4)
image= autoencoder.predict(X_train[300].reshape(28,28,1))
show(image.reshape(28,28))
```

#### Using the trained layers of autoencoder to contrct the Encoder and Decoder networks

```python
encoder_network = Model(input = input_img, output=encoded)

#building the decoder 
encoded_input = Input(shape=(2,))
encoded_layer_1 = autoencoder.layers[-11] # using the trained layers here
encoded_layer_2 = autoencoder.layers[-10]
encoded_layer_3 = autoencoder.layers[-9]
encoded_layer_4 = autoencoder.layers[-8]
encoded_layer_5 = autoencoder.layers[-7]
encoded_layer_6 = autoencoder.layers[-6]
encoded_layer_7 = autoencoder.layers[-5]
encoded_layer_8 = autoencoder.layers[-4]
encoded_layer_9 = autoencoder.layers[-3]
encoded_layer_10 = autoencoder.layers[-2]
encoded_layer_11 = autoencoder.layers[-1]


decoder = encoded_layer_1(encoded_input)
decoder = encoded_layer_2(decoder)
decoder_network = Model(input=encoded_input, output=decoder)
```

### Generating new digits by moving in the latent 2D - space:

Now the hope is that the new 2-D representation of the data is a good coordinate system for the subspace of the data that is actually meaningfull i.e the digits. One way to chwck this is to move in various directions and see if we get some meaningful change in the decoded image of the path or just a noise as we did in the oriignal space?

```python
# moving along the x axis
for i in range(64):
    plt.subplot(8,8,i+1)
    pt = np.array([[i/3,0]])
    show(decoder_network.predict(pt).reshape((28,28)))
    plt.xticks([])
    plt.yticks([])
```

Pretty neat! we see that moving in a given direction in the 2D representation corresponds to staying on the manifolds in the original space. We never end up in static space. Sure not all images are exactly digits but they are all digits llike. You can clearly see that the transformation from one class to another.

Note that the way we are generating images here doesnt have very good statistical properties since we look at the 2D plot first. Using a variational auto encoder makes the generative process more rigorous.

```python
# moving along the y axis
for i in range(64):
    plt.subplot(8,8,i+1)
    pt = np.array([[0,i/3]])
    show(decoder_network.predict(pt).reshape((28,28)))
    plt.xticks([])
    plt.yticks([])
```
